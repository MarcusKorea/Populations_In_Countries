```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_df = pd.read_csv("Data/clean_populations.csv")
df = population_df.loc[:, ["Location", "Time","PopMale", "PopFemale", "PopTotal"]]
df
```

```python
df_sorted = df
df_sorted["percentage_difference_in_gender"] = (df_sorted["PopMale"] - df_sorted["PopFemale"])*100/df_sorted["PopMale"]
df_sorted["percentage_difference_in_gender"] = df_sorted["percentage_difference_in_gender"].abs()
df_sorted = df.sort_values(by = "percentage_difference_in_gender")
df_sorted_lowtohigh = df_sorted
df_sorted_lowtohigh.head(10)
```

```python
df_sorted_hightolow = df_sorted
df_sorted_hightolow = df.sort_values(by = "percentage_difference_in_gender", ascending = False)
df_sorted_hightolow.head(10)
```

```python
df_sorted_hightolow["Location"].unique()
```

```python
difference = []
x=0
y=0
for i in df["PopTotal"]:
    difference.append(x)
    x=((i-y)/i)*100
    y=i
len(difference)
    
```

```python
df["Change"] = difference
df
    
    
```

```python
df1 = df.sort_values(by = "Change" , ascending = False)
df1["Change"] = df1["Change"].round(2)
df1
```
